```python
import numpy as np
import matplotlib.pyplot as plt

# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)

# Sample data
x1 = 3
x2 = 2
y_actual = 17

# Initialize weights
np.random.seed(42)  # For reproducibility
w1 = np.random.uniform(1, 10)
w2 = np.random.uniform(1, 10)
learning_rate = 0.01

# Lists to store error and predictions
errors = []
predictions = []

# Training loop
for epoch in range(20):
    # Feed Forward
    y_pred = x1 * w1 + x2 * w2

    # Error Calculation
    error = (y_actual - y_pred) ** 2
    errors.append(error)
    predictions.append(y_pred)

    # Back Propagation
    grad_w1 = -2 * x1 * (y_actual - y_pred)
    grad_w2 = -2 * x2 * (y_actual - y_pred)

    # Update weights
    w1 -= learning_rate * grad_w1
    w2 -= learning_rate * grad_w2

# Visualization
plt.figure(figsize=(12, 5))

# Error plot
plt.subplot(1, 2, 1)
plt.plot(errors, label='Error')
plt.title('Error over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.legend()

# Predictions plot
plt.subplot(1, 2, 2)
plt.plot(predictions, label='Predictions', color='orange')
plt.axhline(y=y_actual, color='r', linestyle='--', label='Actual Value')
plt.title('Predictions over Epochs')
plt.xlabel('Epochs')
plt.ylabel('Predicted Value')
plt.legend()

plt.tight_layout()
plt.show()
```

Overall, the visualizations indicate that the model is learning effectively, reducing both the error and the difference between predicted and actual values. This is a good outcome, as it reflects the model's ability to generalize well after training.
